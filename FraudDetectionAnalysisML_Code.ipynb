# Loading Data

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
ccd = pd.read_csv('/content/CreditCardFraud/creditcard.csv')
```

```python
ccd.head()
```

```python
ccd.head(2)
```

```python
ccd.info()
```

```python
ccd.isnull().sum()
```

```python
ccd['Class'].value_counts()
```

```python
legit = ccd[ccd.Class == 0]
fraud = ccd[ccd.Class == 1 ]
```

```python
legit.shape
fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
ccd.groupby('Class').mean()
```

```python
legit_samlpe = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_samlpe, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.info()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X.shape
```

```python
Y.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

```python
Y_train.shape
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
pred_Y = model.predict(X_test)
pred_Y
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```
